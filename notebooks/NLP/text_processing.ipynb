### If we use w+ when opening a file all the contents that might contain the file will be erased

```python
my_file = open("file1.txt")
for line in my_file.readlines():
    print(f"This is a line >>>>> {line}")
```

```python
my_file = open("file1.txt", "w+")
my_file.read()
```

```python
my_file.write("New content after opening the file with w+ option")
my_file.seek(0)
my_file.read()
```

```python
my_file.close()
```

```python
my_file = open("whoops.txt", "a+")
my_file.write("Appending text to the new file using A+")
my_file.close()
```

```python
my_file = open("whoops.txt") #only read operation
my_file.read()
```

```python
my_file.write("Should give an error in this line because we opened the file just for reading")
my_file.close()
```

```python
my_file = open("whoops.txt", "a+")
my_file.write("This is an added line using a+")
```

```python
my_file.seek(0)
my_file.read()
```

```python
my_file.write("\n this is a real new line")
my_file.seek(0)
print(my_file.read())
```
